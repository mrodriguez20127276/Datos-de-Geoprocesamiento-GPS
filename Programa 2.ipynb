```python
# Importamos los modulos necesarios 
import gpxpy
import folium
import pandas 
from numpy import *
```

```python
# Abrimos el archivo gpx del recorrido 
gpx = gpxpy.parse(open("20181210.gpx"))
#imprimimos 
gpx
```

```python
print("{} track(s)".format(len(gpx.tracks))) #
track = gpx.tracks[0]

print("{} segment(s)".format(len(track.segments)))
segment = track.segments[0]

print("{} point(s)".format(len(segment.points)))
```

```python
#Creamos una lista vacia para amacenar los datos del archivo gpx
data = []
segment_length = segment.length_3d()
for point_idx, point in enumerate(segment.points):
    data.append([point.longitude, point.latitude,point.elevation, point.time])
    
#imprimimos los datos 
data
```

```python
#Creamos un dataframe para almacenar los datos de interes 
columns = ['Latitude', 'Longitude', 'Altitude', "time"]
df = pandas.DataFrame(data, columns=columns)
```

```python
#Extraemos las coordenadas de Latitude(x) y Longitude(y)
lon = df["Longitude"]
lat = df["Latitude"]
tiempo = df["time"]
```

```python
# Calculamos la distancia entre los puntos de la latitud (x)
dx = []
for i in range(len(lat)-1):
    u= sqrt((-lat[i+1]- lat[i])**2)
    dx.append(u)
    
#Calculamos la distancia entre los puntos de la longitud(y)
dy = []
for i in range(len(lon)-1):
    r= sqrt((-lon[i+1]- lon[i])**2)
    dy.append(r)
```

```python
# Realizamos una sumatoria de las distancias tanto en lat y lon , para calcular la distancia total
dT = []
for i in range(len(dx)):
    dt = dx[i] + dy[i] 
    dT.append(dt)
dT.append(dT[0])

Dist = (sum(dT)/100)/1000
print "Distancia total recorrida = " + str(Dist) + "km" 
```
